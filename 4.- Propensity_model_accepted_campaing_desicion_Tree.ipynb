```python
# Creating the variable with the vector to predict:
y_target = df['Promotion_#_campaign_accepted'].values
df = df.drop(['Promotion_#_campaign_accepted'], axis = 1)
```

```python
# Measuaring the accuracy of the model
model_accuracy = round(model.score(x_test, y_test), 4)
print('Mean accuracy: %0.4f' % (model_accuracy))
```
